```python
X = df.drop("Outcome",axis=1)
y = df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=15)
```

```python
columns_fill = columns_check
```

```python
columns_fill
```

```python
medians = {}

for col in columns_fill:
    median_values = X_train[X_train[col] != 0][col].median()
    medians[col] = median_values
    X_train[col] = X_train[col].replace(0,median_values)

for col in columns_fill:
    X_test[col] = X_test[col].replace(0,medians[col])
```

```python
X_train.describe()
```

```python
X_test.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
ada = AdaBoostClassifier()
ada.fit(X_train,y_train)
y_pred = ada.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
adaboost_params = {
    "n_estimators": [50,70,100,120,150,200],
    "learning_rate": [0.001,0.01,0.1,1,10]
}
```

```python
grid = GridSearchCV(estimator=AdaBoostClassifier(),param_grid=adaboost_params,cv=5,verbose=1,n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
ada = AdaBoostClassifier(learning_rate=1,n_estimators=150)
```

```python
ada.fit(X_train,y_train)
y_pred = ada.predict(X_test)

print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
